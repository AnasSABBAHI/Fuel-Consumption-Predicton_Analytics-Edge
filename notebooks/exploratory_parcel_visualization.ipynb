# Exploratory Parcel Visualization

This notebook provides a simple script to visualize the geographical boundaries of a single agricultural parcel. It reads a parcel's CSV file, which contains latitude and longitude points, and plots its shape.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Parcel Data

We load the data for a single parcel. **Note:** The original hard-coded path has been replaced with a relative path.

```python
# Load the data for a specific parcel
# Replace '41010-20.csv' with another file name to visualize a different parcel.
parcel_df = pd.read_csv('../data/Parcelles/41010-20.csv')
```

## Plot Parcel Boundaries

```python
latitude = parcel_df['Latitude'].tolist()
longitude = parcel_df['Longitude'].tolist()

# Ensure the area is closed by appending the first point at the end
latitude.append(latitude[0])
longitude.append(longitude[0])

# Plot the parcel's circumference
plt.figure(figsize=(10, 8))
plt.plot(longitude, latitude, marker='o', linestyle='-', color='blue', label='Parcel Boundary')
plt.fill(longitude, latitude, color='lightblue', alpha=0.5, label='Parcel Area')

# Add plot details
plt.title('Geographic Visualization of Parcel 41010-20', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend()
plt.grid(True)
plt.axis('equal') # Ensure aspect ratio is equal for a correct geographical representation
plt.show()
```
